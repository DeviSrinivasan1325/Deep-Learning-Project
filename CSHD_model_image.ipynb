```python
# Logic for child and dangerous object position calculation
from ultralytics import YOLO
import pyttsx3
import winsound
import math

model = YOLO("runs/detect/train/weights/best.pt")

results = model("test_images/Baby_with_knife.png", conf=0.25)

child_center = None
danger_center = None

for r in results:
    for box in r.boxes:
        cls = int(box.cls)
        label = model.names[cls]

        x1, y1, x2, y2 = box.xyxy[0]

        center_x = (x1 + x2) / 2
        center_y = (y1 + y2) / 2

        if label == "child":
            child_center = (center_x, center_y)

        if label in ["knife", "scissors"]:
            danger_center = (center_x, center_y)


# Distance calculation
if child_center is not None and danger_center is not None:

    distance = math.sqrt(
        (child_center[0] - danger_center[0])**2 +
        (child_center[1] - danger_center[1])**2
    )

    print("Distance:", distance)

    # Alert if very close
    if distance < 150:
        print("⚠ WARNING: Child near dangerous object!")

        winsound.Beep(1000, 800)

        engine = pyttsx3.init()
        engine.say("Warning! Child near dangerous object")
        engine.runAndWait()

else:
    print("Child or dangerous object not detected")
```
