```python
# Logic for child and dangerous object position calculation
from ultralytics import YOLO
import pyttsx3
import winsound
model = YOLO("runs/detect/train5/weights/best.pt")

results = model("test_images/Baby1.jpg",conf=0.1)

child_center = None
danger_center = None

for r in results:
    for box in r.boxes:
        cls = int(box.cls)
        label = model.names[cls]
    
        x1, y1, x2, y2 = box.xyxy[0]
    
        center_x = (x1 + x2) / 2
        center_y = (y1 + y2) / 2
    
        if label == "child":
            child_center = (center_x, center_y)
    
        if label in ["knife", "scissors"]:
            danger_center = (center_x, center_y)
#Distance calculation between child and dangerous object
import math

if child_center and danger_center:

    distance = math.sqrt(
        (child_center[0] - danger_center[0])**2 +
        (child_center[1] - danger_center[1])**2
    )

    print("Distance:", distance)

#Alert if child and object distance is very close
if distance < 150:
    print("⚠ WARNING: Child near dangerous object!")

    winsound.Beep(1000,800)

    engine = pyttsx3.init()
    engine.say("Warning! Child near dangerous object")
    engine.runAndWait()
```
